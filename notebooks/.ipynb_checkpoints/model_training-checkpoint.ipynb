```python
hotspot_df = df_clean.groupby(
    ['City', 'year', 'month']
).size().reset_index(name='crime_count')

hotspot_df.head()
```

```python
threshold = hotspot_df['crime_count'].quantile(0.90)
threshold
```

```python
hotspot_df['is_hotspot'] = (hotspot_df['crime_count'] >= threshold).astype(int)

hotspot_df.head()
```

```python
hotspot_df['is_hotspot'].value_counts()
```

```python
total_city_months = len(hotspot_df)
total_hotspots = hotspot_df['is_hotspot'].sum()

hotspot_percentage = (total_hotspots / total_city_months) * 100

total_city_months, total_hotspots, hotspot_percentage
```

```python
top_hotspot_cities = (
    hotspot_df[hotspot_df['is_hotspot'] == 1]
    .groupby('City')['crime_count']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

top_hotspot_cities.head(10)
```

```python
monthly_hotspots = (
    hotspot_df[hotspot_df['is_hotspot'] == 1]
    .groupby(['year', 'month'])
    .size()
    .reset_index(name='hotspot_count')
)

monthly_hotspots.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(
    monthly_hotspots['year'].astype(str) + '-' + monthly_hotspots['month'].astype(str),
    monthly_hotspots['hotspot_count']
)
plt.xticks(rotation=90)
plt.title("Monthly Hotspot Trend")
plt.xlabel("Year-Month")
plt.ylabel("Number of Hotspots")
plt.tight_layout()
plt.show()
```

```python
map_hotspot_data = hotspot_df[hotspot_df['is_hotspot'] == 1][
    ['City', 'year', 'month', 'crime_count']
]

map_hotspot_data.head()
```

```python
# sort correctly
hotspot_df = hotspot_df.sort_values(['City', 'year', 'month'])

# previous month's crime count per city
hotspot_df['prev_month_crime'] = (
    hotspot_df.groupby('City')['crime_count']
    .shift(1)
)

hotspot_df.head()
```

```python
hotspot_df['prev_month_crime'] = hotspot_df['prev_month_crime'].fillna(0)
```

```python
hotspot_df['risk_score'] = (
    0.6 * hotspot_df['crime_count'] +
    0.4 * hotspot_df['prev_month_crime']
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

hotspot_df['risk_score_norm'] = scaler.fit_transform(
    hotspot_df[['risk_score']]
)

hotspot_df[['City', 'year', 'month', 'risk_score_norm']].head()
```

```python
def risk_label(score):
    if score >= 0.7:
        return 'High'
    elif score >= 0.4:
        return 'Medium'
    else:
        return 'Low'

hotspot_df['risk_level'] = hotspot_df['risk_score_norm'].apply(risk_label)

hotspot_df[['City', 'risk_score_norm', 'risk_level']].head()
```

```python
predicted_hotspots = hotspot_df[hotspot_df['risk_level'] == 'High']

predicted_hotspots.head()
```

```python
predictive_dashboard_data = (
    predicted_hotspots
    .groupby('City')
    .agg(
        avg_risk=('risk_score_norm', 'mean'),
        hotspot_months=('risk_level', 'count')
    )
    .sort_values(by='avg_risk', ascending=False)
    .reset_index()
)

predictive_dashboard_data.head(10)
```

```python
hotspot_df['alert_flag'] = (hotspot_df['risk_level'] == 'High').astype(int)

hotspot_df[['City', 'year', 'month', 'risk_level', 'alert_flag']].head()
```

```python
def generate_alert(row):
    if row['risk_level'] == 'High':
        return f"High crime risk predicted in {row['City']} for upcoming period."
    elif row['risk_level'] == 'Medium':
        return f"Moderate crime risk observed in {row['City']}."
    else:
        return "Low crime risk."

hotspot_df['alert_message'] = hotspot_df.apply(generate_alert, axis=1)

hotspot_df[['City', 'risk_level', 'alert_message']].head()
```

```python
deployment_priority = (
    hotspot_df[hotspot_df['risk_level'] == 'High']
    .groupby('City')
    .agg(
        avg_risk=('risk_score_norm', 'mean'),
        recent_crime=('crime_count', 'sum')
    )
    .sort_values(by=['avg_risk', 'recent_crime'], ascending=False)
    .reset_index()
)

deployment_priority.head(10)
```

```python
def resource_recommendation(risk):
    if risk >= 0.8:
        return "Increase patrols and surveillance"
    elif risk >= 0.6:
        return "Moderate increase in patrol units"
    else:
        return "Normal patrol"

deployment_priority['recommended_action'] = deployment_priority['avg_risk'].apply(
    resource_recommendation
)

deployment_priority
```
